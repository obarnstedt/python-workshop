```python
a = 10
while a > 0:
    print(a)
    if a > 5:
        print("Big number!")
    elif a % 2 != 0:
        print("This is an odd number")
        print("It isn't greater than five, either")
    else:
        print("this number isn't greater than 5")
        print("nor is it odd")
        print("feeling special?")
    a = a - 1
    print("we just made 'a' one less than what it was!")
    print("and unless a is not greater than 0, we'll do the loop again.")
print("well, it seems as if 'a' is now no bigger than 0!")
print("the loop is now over, and without furthur adue, so is this program!")
```

The code to be executed if the conditions are met MUST BE INDENTED. This is good programming practice in any language, but Python requires that you do it.

# FOR loops

```python
# Example 'for' loop
# First, create a list to loop through:
newList = [45, 'eat me', 90210, "The day has come, the walrus said, \
to speak of many things", -67]

# create the loop:
# Goes through newList, and sequentially puts each bit of information
# into the variable value, and runs the loop
for value in newList:
    print(value)
```

```python
# Loop through dictionary
for key, value in a_dictionary.items():  # iterating over a dictionary
  print(f'key: {key}; value: {value}')
```

# Exercise:
Write a function "melancholic" to write \
"I've counted x raindrops" \
cumulatively for x iterations. Then call the function by providing x with the number 13.

```python
def melancholic(x):
    a = 0
    while a < x:
        print("I've counted {} raindrops".format(a))
        a += 1
```

```python
melancholic(13)
```
